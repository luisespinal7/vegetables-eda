### Ejercicio dos: Por categoria, cuantos distintos items hay y cuantos distintos variety hay?

El resultado debe de tener el siguiente formato de presentacion:

```python
pd.DataFrame({
    "category" : ['cut_flowers','fruit']
    , "item_count" : [9,10]
    , "variety_Count" : [10,21]
})
```

Como nota en un mismo dataframe mostramos todo lo solicitado, usted debe de mostrar todas las categorias

```python
result2 = pd.DataFrame()
##iteramos para obtener las distintas categorías.
for category_ in vdf.loc[ : , "category"].drop_duplicates():
    category_vdf2 = vdf.loc[vdf["category"]==category_, :]
    ##se realiza un conteo de cada item y variety
    item_count_ = category_vdf2["item"].drop_duplicates().count()
    variety_count_ = category_vdf2["variety"].drop_duplicates().count()
    ##se ingresan los datos en un dataframe para luego agregarlos como un registro al dataframe final por medio de una concatenación.
    row2 = pd.DataFrame({
        'category' : [category_],
        'item_count' : [item_count_],
        'variety_count' : [variety_count_]
    })
    
    result2 = pd.concat([result2, row2])
result2
```

### Ejercicio tres: Para la combinacion item, variety muestre el ultimo valor registrado por año.

Ejemplo de como debe de mostrar su respuesta

```python
pd.DataFrame({
    "item" : ['Manzana']
    , "variety" : ['Gala']
    , "year" : [2022]
    , "last_price" :  [3.2]
})
```

```python
result3 = pd.DataFrame()
##iteramos para obtener los valores de los item, variety y año sin repetirse.
for item_ in vdf.loc[ : , "item"].drop_duplicates():
    item_vdf3 = vdf.loc[vdf["item"]==item_, :].copy()
    for variety_ in item_vdf3.loc[ : , "variety"].drop_duplicates():
        variety_vdf3 = item_vdf3.loc[item_vdf3["variety"]==variety_, :].copy()
        
        ##convertimos la columna "date" a tipo datetime para poder obtener solamente el año.
        variety_vdf3["date"] = pd.to_datetime(variety_vdf3["date"])
        
        ##creamos una nueva columna "año" que contenga el año de cada registro.
        variety_vdf3["year"] = variety_vdf3["date"].dt.year
        for year_ in variety_vdf3.loc[ : , "year"].drop_duplicates():
            year_vdf3 = variety_vdf3.loc[variety_vdf3["year"]==year_, :].copy()
            
            ##ordenamos los valores por la columna "date" y creamos una nueva columna "rank" para luego quedarnos solo con el primer valor.
            x3 = year_vdf3.sort_values(by="date")
            x3["rank"]=x3.groupby(["item","variety"], as_index=False)["date"].rank(ascending = False)
            
            ##obtenemos el último valor registrado
            last_price_ = x3.loc[x3["rank"]==1.0, "price"].values
    
            ##creamos un dataframe con el registro de los valores obtenidos para luego concatenarlo con los demás.
            row3 = pd.DataFrame({
                'item' : [item_],
                'variety' : [variety_],
                'year' : [year_],
                'last_price' : [last_price_]
            })
    
            result3 = pd.concat([result3, row3])
result3
```

### Ejercicio cinco: Cual es precio promedio por item, variety

Ejemplo de como se debe de mostrar su respuesta

```python
pd.DataFrame({
    "item" : ['Manzana']
    , "variety" : ['Gala']
    , "year-month" : ['2021-01']   
    , "avg-price" : [0.5]
})
```

```python
result5 = pd.DataFrame()

##iteramos para obtener los valores necesario sin repetirse.
for item_ in vdf.loc[ : , "item"].drop_duplicates():
    item_vdf5 = vdf.loc[ vdf["item"]==item_, : ]
    for variety_ in item_vdf5.loc[ : , "variety"].drop_duplicates():
        variety_vdf5 = item_vdf5.loc[ item_vdf5["variety"]==variety_, : ].copy()
        
        ##pasamos la columna "date" a tipo datetime para poder obtener los años y meses por individual
        variety_vdf5.loc[:,"date"] = pd.to_datetime(variety_vdf5["date"]).copy()
        variety_vdf5.loc[:, "year"] = variety_vdf5["date"].dt.year
        variety_vdf5.loc[:, "month"] = variety_vdf5["date"].dt.month
        
        for year_ in variety_vdf5.loc[ : , "year"].drop_duplicates():
            year_vdf5 = variety_vdf5.loc[variety_vdf5["year"]==year_, : ].copy()
            
            for month_ in year_vdf5.loc[ : , "month"].drop_duplicates():
                month_vdf5 = year_vdf5.loc[year_vdf5["month"]==month_, : ].copy()
                
                ##calculamos el precio promedio.
                avg_price = month_vdf5.groupby(["item", "variety"])["price"].mean().values
                
                ##creamos un dataframe con el registro de los valores obtenidos para luego concatenarlo con los demás.
                row5 = pd.DataFrame({
                    'item': [item_],
                    'variety' : [variety_],
                    'year': [year_],
                    'month': [month_],
                    'avg-price': [avg_price]
                })
                result5 = pd.concat([result5, row5])
result5    
```

### Ejercicio seis: muestre por category,item,variety que tanto % ha aumentado su precio

Ejemplo: si para las manzanas gala el primer precio que se registro fue 2.00 y el ultimo precio que se registro es 4.00 este aumento un 100%

```python
pd.DataFrame({
    "category": ["fruta","fruta"]
    ,"item" : ['Manzana','Manzana']
    , "variety" : ['Gala','egremont_russet']
    , "percentage-change-between-first-and-last-price" : [200,150]       
})
```

```python
result6 = pd.DataFrame()

##iteramos para obtener los valores necesario sin repetirse.
for category_ in vdf.loc[ : , "category"].drop_duplicates():
    category_vdf6 = vdf.loc[ vdf["category"]==category_, : ].copy()
    
    for item_ in category_vdf6.loc[ : , "item"].drop_duplicates():
        item_vdf6 = category_vdf6.loc[ category_vdf6["item"]==item_, : ].copy()
        
        for variety_ in item_vdf6.loc[ : , "variety"].drop_duplicates():
            variety_vdf6 = item_vdf6.loc[ item_vdf6["variety"]==variety_, : ].copy()
            
            ##ordenamos los valores y creamos columnas rank ascendente y descendente para poder obtener el valor más antiguo y el más reciente con un "rank=1.0"
            x6 = variety_vdf6.sort_values(by="date")
            x6["rank_descending"]=x6.groupby(["item","variety"], as_index=False)["date"].rank(ascending = False)
            x6["rank_ascending"]=x6.groupby(["item","variety"], as_index=False)["date"].rank(ascending = True)
            first_value = x6.loc[x6["rank_ascending"]==1.0, "price"].copy().values
            last_value = x6.loc[x6["rank_descending"]==1.0, "price"].copy().values
            
            ##calculamos el porcentaje de aumento del úlltimo valor respecto del primero
            percentage_change = ((last_value-first_value)/first_value)*100
            
            ##creamos un dataframe con el registro de los valores obtenidos para luego concatenarlo con los demás.
            row6 = pd.DataFrame({
                'category' : [category_],
                'item': [item_],
                'variety' : [variety_],
                'percentage-change-between-first-and-last-price': [percentage_change]
            })
            result6 = pd.concat([result6, row6])
result6   
```
